```python
import pandas as pd 
import numpy as np 
import requests
import os

import pypyodbc as odbc #pip install pypyodbc

# Database connection details
DRIVER = "SQL Server"
SERVER_NAME = "DESKTOP-FB9GSJB"
DATABASE_NAME = "EnergyConsumption"
```

1. Download & Get "Building Detail" data

```python
# Function to download the CSV file
def download_BuildingDetails():
    csv_url = "https://services2.arcgis.com/11XBiaBYA9Ep0yNJ/arcgis/rest/services/BuildingDetails/FeatureServer/replicafilescache/BuildingDetails_-7832066962313449791.csv"

    # Download CSV
    response = requests.get(csv_url)

    if response.status_code == 200:
        file_path = os.path.abspath("Building_Details.csv")
        with open(file_path, "wb") as file:
            file.write(response.content)
        print("CSV file downloaded successfully!")
    else:
        print("Failed to download CSV file.")
        return None

    # Load CSV into DataFrame
    building_detail = pd.read_csv(file_path, dtype={"Building_Name": "str"}, low_memory=False)
    return building_detail

building_detail = download_BuildingDetails()
building_detail
```

2. Download & Get " HRM Building Energy Consumption" Data

```python
def download_BuildingEnergyConsumption():

    # Direct URL to the CSV file
    csv_url = "https://services2.arcgis.com/11XBiaBYA9Ep0yNJ/arcgis/rest/services/HRM_Building_Energy_Usage/FeatureServer/replicafilescache/HRM_Building_Energy_Usage_-3651640421373328302.csv"

    # Send a GET request to the URL
    response = requests.get(csv_url)

    # Check if the request was successful
    if response.status_code == 200:
        # Write the content to a CSV file
        with open("HRM_Building_Energy_Usage.csv", "wb") as file:
            file.write(response.content)
        print("CSV file downloaded successfully!")
    else:
        print("Failed to download CSV file.")

    file_path = r"HRM_Building_Energy_Usage.csv"

    energy_usage = pd.read_csv(file_path)

    return energy_usage

energy_usage = download_BuildingEnergyConsumption()
energy_usage
```

3. Manipulate Data

```python
# Function to clean and preprocess the data
def manipulate_BuildingData(building_detail):
    # Remove duplicates
    building_detail = building_detail.drop_duplicates()

    # Filter for buildings where PRIMARY_USE is "Y"
    if "PRIMARY_USE" in building_detail.columns:
        building_detail = building_detail[building_detail["PRIMARY_USE"] == "Y"]

    # Select only relevant columns
    selected_columns = [
        "BUILDING_ID", "BUILDING_NAME", "YEAR_OF_CONSTRUCTION",
        "TOTAL_SQUARE_FOOTAGE", "BUILDING_CLASSIFICATION", "BUILDING_USE"
    ]
    building_detail = building_detail[selected_columns]

    # Rename columns to match SQL database
    building_detail = building_detail.rename(columns={
        "BUILDING_ID": "Building_ID",
        "BUILDING_NAME": "Building_Name",
        "YEAR_OF_CONSTRUCTION": "Year_Built",
        "TOTAL_SQUARE_FOOTAGE": "Total_Sqft",
        "BUILDING_CLASSIFICATION": "Building_Classification",
        "BUILDING_USE": "Building_Use"
    })

    building_detail = building_detail.dropna(subset=["Year_Built"])

    return building_detail


building = manipulate_BuildingData(building_detail)
building

```

```python
# Function to create a connection string
def connection_string(driver, server_name, database_name):
    return f"""
        DRIVER={{{driver}}};
        SERVER={server_name};
        DATABASE={database_name};
        Trusted_Connection=yes;
    """
```

```python
# Function to insert data into SQL Server
def insert_into_sql(building):
    # Ensure Building_ID is a string and clean it
    building["Building_ID"] = building["Building_ID"].astype(str).str.strip()
    building["Building_ID"] = building["Building_ID"].str.replace(r"[^\x00-\x7F]+", "", regex=True)
    # Remove duplicate Building_ID before inserting
    building = building.drop_duplicates(subset=["Building_ID"], keep="first")

    # Clean Building_Name (fix NaN issue)
    building["Building_Name"] = building["Building_Name"].astype(str).fillna("Unknown")

    # Clean text fields and replace special characters
    building["Building_Classification"] = building["Building_Classification"].astype(str).fillna("Unknown").str.replace("/", "-", regex=True)
    building["Building_Use"] = building["Building_Use"].astype(str).fillna("Unknown").str.replace("/", "-", regex=True)

    # Convert numerical columns
    building["Year_Built"] = pd.to_numeric(building["Year_Built"], errors="coerce").fillna(0).astype(int)
    building["Total_Sqft"] = pd.to_numeric(building["Total_Sqft"], errors="coerce").fillna(0).astype(float)

    # Select only required columns
    columns = ["Building_ID", "Building_Classification", "Building_Use", "Building_Name", "Year_Built", "Total_Sqft"]
    df_data = building[columns]

    # Print first few rows to debug before inserting
    print("\n✅ Final Data Preview Before Insert:")
    print(df_data.dtypes)
    print(df_data.head(10))
    
    # Convert DataFrame to list of lists
    records = df_data.values.tolist()
    
    print(f"✅ Total Records to Insert: {len(records)}")
    print("✅ Sample Data:", records[:5])  # Show first 5 records

    # Connect to SQL Server
    try:
        conn = odbc.connect(connection_string(DRIVER, SERVER_NAME, DATABASE_NAME))
        conn.autocommit = True  
        cursor = conn.cursor()
        print("\n✅ Connected to SQL Server successfully!")
    except Exception as e:
        print("\n❌ Connection Error:", str(e))
        return

    # SQL Query to insert data
    sql_insert = """
        INSERT INTO Building (Building_ID, Building_Classification, Building_Use, Building_Name, Year_Built, Total_Sqft) 
        VALUES (?, ?, ?, ?, ?, ?)
    """

    # Insert data into SQL Server
    try:
        if len(records) > 0:  # Only attempt insert if records exist
            cursor.executemany(sql_insert, records)
            conn.commit()
            print("\n✅ Data inserted successfully!")
        else:
            print("\n⚠️ No data to insert.")
    except Exception as e:
        conn.rollback()
        print("\n❌ Error inserting data:", str(e))
    finally:
        cursor.close()
        conn.close()
        print("\n✅ Connection closed.")

# Main Execution
if __name__ == "__main__":
    building_detail = download_BuildingDetails()
    if building_detail is not None:
        building = manipulate_BuildingData(building_detail)
        insert_into_sql(building)

# Final Check
print("\n✅ Final Data Preview Before Insert:")
print(building.dtypes)  # Check data types
print(building.head(10))  # Preview first 10 rows
print("\n✅ Checking for NaN values:\n", building.isna().sum())
```

```python
def manipulate_EnergyUsageData(energy_usage):

    # Remove duplicates
    energy_usage = energy_usage.drop_duplicates()

    # Select relevant columns
    selected_columns = [
        "Energy Type", "HRM Building ID", "Start Date", "End Date",
        "Consumption", "Unit of Measure", "Cost"
    ]
    
     # Ensure only selected columns remain
    energy_usage = energy_usage.loc[:, selected_columns]
    
    energy_usage.rename(columns = {"HRM Building ID": "BUILDING_ID"}, inplace=True)
    
    return energy_usage

building_energy = manipulate_EnergyUsageData(energy_usage)
building_energy

```
